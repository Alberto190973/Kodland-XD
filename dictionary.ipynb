```python
meme_dict = {
            "CRINGE": "Algo excepcionalmente raro o embarazoso",
            "LOL": "En ingles Significa LAUGH OUT LOUD, en español es reír en voz alta",
            "LIL_BRO": "Un adjetivo que significa que eres pequeño",
            "SIGMA": "Se refiere a una persona que es atractiva",
            "XD": "Una respuesta común a algo gracioso"
            "BRAINROT": "algo que te pudre el cerebro"
            }

def Find_Word(word):
    if word in meme_dict.keys():
        print(meme_dict[word])
    else:
        print("No tenemos esa palabra todavia!")

word = input("Escribe una palabra que no entiendas (¡con mayúsculas!): ")
Find_Word(word)
```
